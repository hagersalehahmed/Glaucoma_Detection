```python
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
original_dataset_dir = 'dataset'  # Path to the dataset with subfolders
base_dir = 'new_directories1'  # Where you want to save new directories

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')
```

```python
for dir in [train_dir, validation_dir, test_dir]:
    os.makedirs(dir, exist_ok=True)
```

```python
# Split ratio
train_ratio = 0.7
validation_ratio = 0.1
test_ratio = 0.2

# Function to split and copy files
def split_and_copy_files(class_dir, dest_base_dir):
    all_images = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if os.path.isfile(os.path.join(class_dir, f))]
    train_images, test_images = train_test_split(all_images, test_size=1-train_ratio, random_state=42)
    validation_images, test_images = train_test_split(test_images, test_size=test_ratio/(test_ratio + validation_ratio), random_state=42)

    for images, folder_name in zip([train_images, validation_images, test_images], ['train', 'validation', 'test']):
        dest_dir = os.path.join(dest_base_dir, folder_name, os.path.basename(class_dir))
        os.makedirs(dest_dir, exist_ok=True)
        for file in images:
            shutil.copy(file, dest_dir)

# Iterate over each class directory and split the files
for class_dir in os.listdir(original_dataset_dir):
    full_class_dir = os.path.join(original_dataset_dir, class_dir)
    if os.path.isdir(full_class_dir):
        split_and_copy_files(full_class_dir, base_dir)
```
